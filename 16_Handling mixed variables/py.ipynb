```python
# extract numerical part
df['number_numerical'] = df['number'].str.extract(r'(\d+)').astype(float)
```

```python
#extract categorical part
df['number_categorical'] = df['number'].str.extract(r'([a-zA-Z]+)')
```

```python
df['cabin_num']=df['Cabin'].str.extract(r'(\d+)').astype(float) #capture the numerical part of the cabin number
df['cabin_cat']=df['Cabin'].str[0] #capture the first letter of the cabin number

df.head()
```

```python
df[df['Ticket'].str.extract(r'(\d+)$')[0].isna()]
```

```python
#extract last bit of ticket as number
#Ticket column contains two pieces of information:
#Ticket Category (Prefix) → A/5, PC, STON/O2., PP
#Ticket Number → 21171, 17599, 3101282, 113803
df['Ticket_Number'] = (
    df['Ticket']
      .str.extract(r'(\d+)$')[0]
      .astype('Int64')
)
df['ticket_category'] = (
    df['Ticket']
      .str.extract(r'^(.*?)(?=\s+\d+$)')[0]
)

df.head(20)
```
