## converters


```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
#it will convert all Royal Challengers Bangalore to RCB in team1 col
```

## na_values parameters


```python
# soemtimes you have some other na values lets say '~' or 'na' or 'n/a' or 'not available'
# so we can use na_values param to convert them to NaN
pd.read_csv('aug_train.csv',na_values=['~','na','n/a','not available'])
```
